```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_excel("C:/Users/Admin/Downloads/titanic3.xls")
print(df)
```

```python
# Describe Dataset
print(df.describe())
```

```python
# Dataset Info
print(df.info())
```

```python
# Dataset value count
print(df.value_counts())
```

```python
# Using sns Pairplot
sns.pairplot(df[['age','fare','pclass','survived']].dropna(),hue='survived')
plt.suptitle("Pairplot of Iris Dataset ", y=1.02)
plt.show()
```

```python
# Seaborn Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True,cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

```python
# Seaborn Histograms
df[['age','fare','pclass']].hist(figsize=(10,6),edgecolor='black')
plt.suptitle("Histograms of age, fare, and pclass")
plt.tight_layout()
plt.show()
```

```python
# Seaborn Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='survived',y='age',data=df)
plt.title("Boxplot of age by survival")
plt.show()
```
